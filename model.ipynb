```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns
from sklearn.model_selection import train_test_split
#import lightgbm as lgb
from sklearn.ensemble import RandomForestRegressor
#from xgboost import XGBRegressor
from sklearn.metrics import r2_score, accuracy_score,mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
data_train = pd.read_csv("train.csv")
data_test = pd.read_csv("test.csv")
```

```python
data_train = data_train.drop(['id'], axis = 1)
data_test = data_test.drop(['id'], axis = 1)
```

```python
data_train_copy = data_train.copy()
X = data_train_copy.drop(columns='pressure')
y = data_train_copy['pressure']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3,random_state = 1)
```

```python
from catboost import CatBoostRegressor
cbr = CatBoostRegressor(learning_rate=0.1, max_depth=10, n_estimators=900, subsample=0.5)
cbr.fit(X_test, Y_test)
```

```python
def mape(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.mean(np.abs((actual - pred) / actual)) * 100
```

```python
CBR_train_predict = cbr.predict(X_train)
CBR_validation_predict = cbr.predict(X_test)
CBR_R2_train = r2_score(Y_train,CBR_train_predict)
CBR_R2_validation = r2_score(Y_test,CBR_validation_predict)

CBR_MAPE_train = mape(Y_train,CBR_train_predict)
CBR_MAPE_validation = mape(Y_test,CBR_validation_predict)


print("R Squared Score : ", CBR_R2_validation)
print("Mape Score : ", CBR_MAPE_validation)



score = cbr.score(X_test, Y_test)
score
```
